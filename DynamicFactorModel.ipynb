```python
import pandas as pd
data = pd.read_csv("百度搜索指数2017-2023全年-20240112更新.csv")
data
```

```python
data = data.fillna(0)
full_data = data.iloc[:,1:]
```

```python
u_data=(full_data-full_data.min())/(full_data.max()-full_data.min())
#u_data = u_data.rename(columns=full_data.columns)
u_data.head(3)
```

```python
import matplotlib as mpl
 
    
mpl.rcParams['font.sans-serif'] = ['SimHei']  # 指定默认字体
mpl.rcParams['axes.unicode_minus'] = False  # 解决保存图像是负号'-'显示为方块的问题
```

# DFM

```python
factors = dict()
for macro_variable in list(full_data.columns):
    factors[macro_variable] = ["0"]
```

```python
import statsmodels.api as sm
factor_model = sm.tsa.DynamicFactorMQ(u_data,
                    factors=factors,
                    factor_orders = {'0': 1},
                    idiosyncratic_ar1=True,
                    standardize=False)
```

```python
model_results = factor_model.fit(disp=50)
```

```python
model_results.factors.smoothed["0"].values
```
